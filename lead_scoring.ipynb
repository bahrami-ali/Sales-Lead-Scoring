```python
import pandas as pd
from pathlib import Path
import os
data_path = Path(os.getcwd(), "data")
```

```python
df = pd.read_csv(data_path/"Leads.csv")
df.head()
```

```python
df.shape
```

```python
# Rename columns: remove leading/trailing spaces, make it lowercase, and separate it by underscores
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
df_dict = pd.read_excel(data_path/"Leads Data Dictionary.xlsx", header=2, usecols=[1,2])
df_dict.loc[:,"Variables"] = df_dict.loc[:,"Variables"].str.strip().str.lower().str.replace(" ", "_")
df_dict
```
